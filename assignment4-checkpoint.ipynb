```python
import pandas as pd

df = pd.read_csv(r"C:\Users\SOMA\Downloads\Aeka Assignments\Assignment 4\Top Indian Places to Visit.csv")

# Rename columns for easier handling
df = df.rename(columns={
    'Google review rating': 'rating',
    'Number of google review in lakhs': 'popularity',
    'Time needed to visit in hrs': 'visit_time'
})

# Remove rows with missing values
df = df.dropna(subset=['City', 'Name', 'rating', 'popularity', 'time needed to visit in hrs'])

df.head()
```

```python
df['rating_score'] = df['rating'] / df['rating'].max()
df['popularity_score'] = df['popularity'] / df['popularity'].max()
df['visit_time_score'] = 1 - (df['time needed to visit in hrs'] / df['time needed to visit in hrs'].max())
```

```python
df['final_score'] = (
    0.4 * df['rating_score'] +
    0.4 * df['popularity_score'] +
    0.2 * df['visit_time_score']
)
```

```python
def recommend_weekend_getaways(source_city, top_n=5):
    city_data = df[df['City'].str.lower() == source_city.lower()]
    
    ranked = city_data.sort_values(by='final_score', ascending=False)
    
    return ranked[['Name', 'Type', 'rating', 'popularity', 'time needed to visit in hrs']].head(top_n)
```

```python
print("Top weekend places from Kolkata:")
print(recommend_weekend_getaways("Kolkata"))

print("\nTop weekend places from Delhi:")
print(recommend_weekend_getaways("Delhi"))

print("\nTop weekend places from Mumbai:")
print(recommend_weekend_getaways("Mumbai"))
```
